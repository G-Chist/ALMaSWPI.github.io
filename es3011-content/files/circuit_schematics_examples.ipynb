```python
# Parallel RC
with schemdraw.Drawing() as d:
    elm.SourceV().label('V').right()
    elm.Dot()
    d.push()
    elm.Resistor().label('R').up()
    elm.Line().right()
    elm.Ground()
    d.pop()
    elm.Capacitor().label('C').right()
    elm.Ground()
```

```python
# Parallel LC
with schemdraw.Drawing() as d:
    elm.SourceV().label('V').right()
    elm.Dot()
    d.push()
    elm.Inductor().label('L').up()
    elm.Line().right()
    elm.Ground()
    d.pop()
    elm.Capacitor().label('C').right()
    elm.Ground()
```

```python
# Parallel RLC
with schemdraw.Drawing() as d:
    elm.SourceV().label('V').right()
    elm.Dot()
    d.push()
    elm.Inductor().label('L').up()
    elm.Line().right()
    elm.Ground()
    d.pop()
    d.push()
    elm.Capacitor().label('C').right()
    elm.Ground()
    d.pop()
    elm.Resistor().label('R').down()
    elm.Ground()
```
