# SymPy Basics

SymPy is a Python library for symbolic mathematics. It provides:
- Symbolic variables and expressions (exact, not floating-point)
- Algebraic solving, calculus, and equation manipulation
- Symbolic linear algebra, Laplace transforms, and ODE solving

```python
import sympy as sp
print(f"SymPy version: {sp.__version__}")
```

## 1. Symbols and Symbolic Expressions

```python
# Define symbols
x, y, z = sp.symbols('x y z')

# Build expressions
expr = x**2 + 2*x + 1
print("Expression:", expr)

# Expanded form
expr2 = (x + y)**3
print("Expanded:", sp.expand(expr2))

# Rational numbers stay exact
print("Rational:", sp.Rational(1, 3) + sp.Rational(1, 6))
```

## 2. Substitution and Numeric Evaluation

```python
expr = x**2 + 2*x + 1

# Substitute a value for x
print("At x=3:", expr.subs(x, 3))

# Multiple substitutions
expr2 = x**2 + y**2
print("At x=2, y=3:", expr2.subs({x: 2, y: 3}))

# Numeric evaluation (float)
pi_expr = sp.pi + sp.sqrt(2)
print("Symbolic:", pi_expr)
print("Numeric:", pi_expr.evalf())
print("With 50 digits:", pi_expr.evalf(50))
```

## 3. Solving Algebraic Equations

```python
# Solve x**2 - 4 = 0
solutions = sp.solve(x**2 - 4, x)
print("x^2 - 4 = 0 ->", solutions)

# System of equations
sol = sp.solve([x + y - 5, x - y - 1], [x, y])
print("x + y = 5, x - y = 1 ->", sol)

# Quadratic formula (symbolic coefficients)
a, b, c = sp.symbols('a b c')
roots = sp.solve(a*x**2 + b*x + c, x)
print("Quadratic formula:", roots)
```

## 4. Calculus: Differentiation

```python
expr = x**3 * sp.sin(x)
print("Derivative:", sp.diff(expr, x))

# Higher-order derivative (2nd)
print("2nd derivative:", sp.diff(expr, x, 2))

# Partial derivatives
f = x**2 * y + y**3
print("df/dx:", sp.diff(f, x))
print("df/dy:", sp.diff(f, y))
```

## 5. Calculus: Integration

```python
# Indefinite integral
print("∫x^2 dx =", sp.integrate(x**2, x))

# Definite integral
print("∫_0^1 x^2 dx =", sp.integrate(x**2, (x, 0, 1)))

# Improper integral
print("∫_0^∞ e^{-x} dx =", sp.integrate(sp.exp(-x), (x, 0, sp.oo)))
```

## 6. Limits and Series

```python
# Limit
print("lim_{x→0} sin(x)/x =", sp.limit(sp.sin(x)/x, x, 0))

# Series expansion (around x=0, 5 terms)
expr = sp.sin(x)
print("sin(x) series:", sp.series(expr, x, 0, 5).removeO())
```

## 7. Symbolic Matrices

```python
A = sp.Matrix([[1, 2], [3, 4]])
print("Matrix A:")
sp.pprint(A)

print("\nDeterminant:", A.det())
print("Inverse:")
sp.pprint(A.inv())
print("\nEigenvalues:", A.eigenvals())

# Symbolic matrix with symbolic entries
a, b, c, d = sp.symbols('a b c d')
B = sp.Matrix([[a, b], [c, d]])
print("\nSymbolic matrix B:")
sp.pprint(B)
print("\nSymbolic determinant:")
sp.pprint(sp.factor(B.det()))
```

## 8. Simplification and Manipulation

```python
expr = sp.sin(x)**2 + sp.cos(x)**2
print("Simplify trig:", sp.simplify(expr))

expr2 = (x**3 - 1) / (x - 1)
print("Simplify rational:", sp.simplify(expr2))

# Factor and expand
poly = x**2 - 5*x + 6
print("Factor:", sp.factor(poly))
print("Expand back:", sp.expand(sp.factor(poly)))
```

## 9. Laplace and Fourier Transforms

Laplace transforms are central to control engineering.

```python
t, s = sp.symbols('t s')

# Laplace transform of common functions
print("L{1}:", sp.laplace_transform(1, t, s))
print("L{t}:", sp.laplace_transform(t, t, s))
print("L{exp(-a*t)}:", sp.laplace_transform(sp.exp(-a*t), t, s))
print("L{sin(ωt)}:", sp.laplace_transform(sp.sin(b*t), t, s))

# Inverse Laplace
F = 1 / (s*(s + 1))
print("\nInverse L{1/(s(s+1))}:", sp.inverse_laplace_transform(F, s, t))
```

## 10. Ordinary Differential Equations

```python
# Define function and differential equation
f = sp.Function('f')
ode = sp.Eq(f(t).diff(t, t) + 5*f(t).diff(t) + 6*f(t), 0)
print("ODE:", ode)

# Solve ODE
sol = sp.dsolve(ode, f(t))
print("General solution:", sol)

# With initial conditions: f(0)=1, f'(0)=0
ics = {f(0): 1, sp.diff(f(t), t).subs(t, 0): 0}
sol_ivp = sp.dsolve(ode, f(t), ics=ics)
print("With ICs:", sol_ivp)
```
